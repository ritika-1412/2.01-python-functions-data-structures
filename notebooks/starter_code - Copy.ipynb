1. Why modular code matters.

As your code gets longer, copy-pasting the same logic becomes difficult to maintain. Functions solve that problem.

````python
quiz_scores = [80, 90, 100]
````

````python
quiz_average = sum(quiz_scores) / len(quiz_scores)
````

````python
print(quiz_average)
````

````python
exam_scores = [70, 85, 95]
````

````python
exam_average = sum(exam_scores) / len(exam_scores)
print(exam_average)
````

Let's define a function so we can reuse the same code

`def` is the built-in Python keyword that tells Python we're making a function

````python
def calculate_average(scores):
    average = sum(scores) / len(scores)
    return average
quiz_average = calculate_average(quiz_scores)
print(quiz_average) 
exam_average = calculate_average(exam_scores)
print(exam_average)
````

````python
calculate_average(quiz_scores) == quiz_average
````

````python
# see our function in action with our quiz and exam scores

````

Now the logic is written once and reused many times.

Why this matters in real work
Functions help you:

* reduce repetition
* debug more easily
* test one piece of logic at a time
* make code easier for teammates to read
* scale workflows across multiple datasets

### A. Functions Should Usually Do One Clear Thing

````python
def calculate_total(price,quantity):
    total = price * quantity
    return total

calculate_total(100,2)
````

### B. Default Parameters

If we don't specify an argument for a parameter, the default parameter is the argument

````python
def power_number(base=5, exponent=2):
    result = base ** exponent
    return result

power_number(10)
````

````python
power_number(10, 3)
````

````python
power_number()

#the first parameter cant be default but the second one can be. If we want to use the default value for the second parameter, we have to provide a value for the first parameter.
````

### C. Returning vs Printing

`return` sends a value back so it can be reused later.

````python
def add_numbers(num1, num2):
    result = num1 + num2
    return result
x = add_numbers(5, 10)
````

````python
def add_numbers_print(num1, num2):
    result = num1 + num2
    print(result)

    y=add_numbers_print(5, 10)
````

## 6. Validation Inside Functions

Validation checks help catch problems early.

````python
calculate_average(scores)
````

````python
calculate_average(student)
````

````python
def calculate_average(values): 
    if not isinstance(values, list):
        raise TypeError("Input must be a list of numbers.")
    return sum(values) / len(values)
````

````python
calculate_average([])
````

````python
len([])
````

````python
isinstance(student,dict)

# checks if a variable is a certain data tupe then returns true or false
````

````python
help(isinstance)
````

````python
def calculate_average(values): 
    if not isinstance(values, list):
        raise TypeError("Input must be a list of numbers.")
    if not values:
        raise ValueError("Input list cannot be empty.")
    return sum(values) / len(values)
````

### Why validation matters

Validation helps you:

- catch bad inputs early
- produce clearer error messages
- make functions more reliable
- prevent silent mistakes in analysis

## 7. Intro to Pandas for Function-Based Data Cleaning

Pandas is a Python library for working with tabular data.


````python
# import pandas, with the alias
import pandas as pd
````

````python
# define the dictionary
boring_dict = {
    "name": ["Ava", "Luis", "Mina", "Jordan"],
    "score": [95, None, 88, 91],
    "hours_studied": [10, 7, None, 8]
}
````

````python
# now convert it to a dataframe
df = pd.DataFrame(boring_dict)
print(df)
````

````python
# here are a few pandas methods to clean some data, then we'll put in a function
df.dropna() # drops rows with missing values
````

````python
# .dropna() drops ANY missing values from a dataframe
# BE AWARE, YOU CAN LOSE A LOT OF DATA WITH THIS
df

#df = df.dropna() # this will drop the rows with missing values and save the cleaned dataframe as df
````

````python
# we can fill missing values with .fillna()
#to access individual columns we can use df["column_name"]
df["score"]
````

````python
df["score"].fillna(df["score"].mean())

# took score column , told pandas to fill in any missing data with average score
````

````python
# Add a dervided column
df["passed"] = 0
df
````

````python
df["passed"] = df["score"] >= 70
df
````

````python
# Now we'll put them all into functions
def fill_missing_values(df):
    df["score"] = df["score"].fillna(df["score"].mean())
    return df

printdf = fill_missing_values(df)
printdf
````
