## Task 1. Validate Records

Requirements:

* Input must be a list
* Each item must be a dictionary
* Raise appropriate errors if invalid

````python
def validate_records(records):
    # Check if input type is a list
    if not isinstance(records, list):
        raise TypeError("Input must be a list")

    #Check if each item in the list is a dictionary
    for record in records:
        if not isinstance(record, dict):
            raise TypeError("Each item in the listmust be a dictionary")
        
````

````python
## Test the function with input as a list of dictionaries
records = [
    {"name": "Ava", "score": 95, "hours_studied": 10},
    {"name": "Ben", "score": None, "hours_studied": 8},
    {"name": "Cara", "score": 88, "hours_studied": None},
    {"name": "", "score": 72, "hours_studied": 6},
    {"name": "Eli", "score": 105, "hours_studied": 9},
    {"name": "Finn", "score": 65, "hours_studied": 7}
]

output =validate_records(records)
print(output)
````

````python
## Test the function with input as a string
output = validate_records("not a list")
print(output)
````

````python
## Test the function with input as an empty list
output = validate_records([])
print(output)
````

````python
## Test the function with input as list of items that are not dictionaries
input = [0, 1, 2]
output = validate_records(input)
print(output)
````

## Task 2. Clean Names

Requirements:

* Remove records where name is empty ("")
* Return cleaned list

````python
def clean_names(records):
    # remove records with empty names
    for record in records:
        if record["name"] == "":
            records.remove(record)
    return records
````

## Task 4. Fix Invalid Scores

Requirements:

* Ensure all scores are between 0 and 100
* Any score above 100 should be set to 100

````python
def cap_scores(records):
    # Ensure scores are between 0 and 100
    for record in records:
        if record["score"] is not None:
            if record["score"] > 100:
                record["score"] = 100
            elif record["score"] < 0:
                record["score"] = 0
    return records
````

````python
def to_dataframe(records):
    import pandas as pd
    # Convert cleaned records into a Pandas DataFrame
    return pd.DataFrame(records)
````

## Task 6. Full Pipeline

Requirements:

* Call all previous functions in order
* Return final cleaned DataFrame

````python
def clean_dataset(records):
    # Call all functions in order
    validate_records(records)
    clean_names(records)
    fill_missing_values(records)
    cap_scores(records)
    df = to_dataframe(records)
    return df
````
